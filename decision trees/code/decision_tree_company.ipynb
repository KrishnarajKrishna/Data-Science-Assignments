```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets  
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
company = pd.read_csv('Company_Data.csv')
company
```

```python
company.info()
```

```python
company['Sales'].min()
company['Sales'].max()
company['Sales'].value_counts()
```

```python
np.median(company["Sales"])
```

```python

company["sales"]= "<=7.49"
company.loc[company["Sales"]>=7.49,"sales"]=">=7.49"
```

```python
company["sales"].unique()
company["sales"].value_counts()
```

```python

company.replace({'sales':{'>=7.49':1,'<=7.49':0}}, inplace=True)
company
```

```python
company.info()
```

```python
#scaling
label_encoder = preprocessing.LabelEncoder()
company['ShelveLoc'] = label_encoder.fit_transform(company['ShelveLoc'])
company['Urban'] = label_encoder.fit_transform(company['Urban'])
company['US'] = label_encoder.fit_transform(company['US'])
```

```python
company
```

```python
x= company.iloc[:,0:10]
y= company.iloc[:,10]
```

```python
x
```

```python
y
```

```python
col_names = list(company.columns)
col_names
```

```python

#splitting data into training and test set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state= 40)
```

```python
x_train
```

```python
y_train
```

### Building decision tress classifier using entropy criteria


```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#plot decision tree
fig = plt.figure(figsize=(125,75))
tree.plot_tree(model);
```

```python
fn=['CompPrice',
 'Income',
 'Advertising',
 'Population',
 'Price',
 'ShelveLoc',
 'Age',
 'Education',
 'Urban',
 'US']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (20,20), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               filled = True);
```

```python
#predicting on test data
predict_test = model.predict(x_test) 
# count of each category
pd.Series(predict_test).value_counts()  
```

```python
predict_test
```

```python
pd.crosstab(y_test,predict_test) 
## 2 way table to understand the correct and wrong predictions
```
